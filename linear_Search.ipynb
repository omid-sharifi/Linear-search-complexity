```python
import numpy as np

# Define the CSV file path
csv_file_path = 'random_numbers.csv'

# Load data from CSV file into a NumPy array
data_array = np.genfromtxt(csv_file_path, delimiter=',')

# Print the shape of the array
print("Shape of the array:", data_array.shape)

# Print the array
print("Data in the array:\n", data_array)
```

Liear search function:

```python
import numpy as np

def linear_search(arr, target):
    """
    Perform linear search to find the target element in the NumPy array.

    Parameters:
    - arr: The NumPy array to be searched.
    - target: The element to be searched for.

    Returns:
    - index: The index of the target element if found, otherwise -1.
    """
    for i in range(len(arr)):
        if arr[i] == target:
            return i  # Return the index if target is found
    return -1  # Return -1 if target is not found
```

```python
import time
import numpy as np

# Example usage:
allSearchedElements = []
element_CPU_SearchedTime = []

for i in range(30):
    # Record the start time
    #start_time = time.time()
    start_time = time.process_time()

    arr = data_array
    target = data_array[i*17]

    allSearchedElements = np.append(allSearchedElements, target)

    index = linear_search(arr, target)

    #if index != -1:
    #print(f"Element {target} found at index {index}.")
    #else:
     #  print(f"Element {target} not found in the array.")

    # Record the end time
    #end_time = time.time()
    end_time = time.process_time()

    # Calculate the elapsed time
    elapsed_time = end_time - start_time

    element_CPU_SearchedTime = np.append(element_CPU_SearchedTime, elapsed_time)


print(allSearchedElements)

print(element_CPU_SearchedTime)
```

```python
import matplotlib.pyplot as plt

# Sample data and corresponding times
times = element_CPU_SearchedTime # Example times
data = allSearchedElements  # Example data corresponding to times

# Create the plot
plt.plot( data, times,  linestyle='-')

# Add labels and title
plt.xlabel('Data')
plt.ylabel('Time')
plt.title('Data vs Time')

# Show the plot
plt.grid(True)  # Add grid lines
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Sample data
x_data = allSearchedElements
y_data = element_CPU_SearchedTime

# Define the model function
def model_function(x, a, b):
    return a * x + b

# Fit the data to the model
popt, pcov = curve_fit(model_function, x_data, y_data)

# Extract optimized parameters
a_opt, b_opt = popt

# Plot the original data
plt.scatter(x_data, y_data, label='Original Data')

# Plot the fitted curve
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = model_function(x_fit, a_opt, b_opt)
plt.plot(x_fit, y_fit, 'r-', label='Fitted Curve')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Curve Fitting Example')
plt.legend()

# Show the plot
plt.show()
```
